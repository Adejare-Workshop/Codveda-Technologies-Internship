# Task 1 – Data Preprocessing for Machine Learning

Before feeding data into any model, we need to clean and prepare it. This notebook walks through the standard preprocessing steps on the Iris dataset: checking for missing values, encoding labels, normalising features, and splitting into train/test sets.

**Dataset:** Iris (built-in via seaborn — no local file needed)  
**Steps covered:** missing-value check → label encoding → Min-Max normalisation → train/test split

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split
```

##  Load the Dataset

```python
# Using seaborn's built-in copy of Iris so the notebook runs anywhere
iris = sns.load_dataset('iris')

print(f"Shape: {iris.shape}")
iris.head()
```

##  Explore & Check for Missing Values

```python
# A quick look at column types and non-null counts
iris.info()
```

```python
# Check if any column has nulls — ideally all zeros here
print("Missing values per column:")
print(iris.isnull().sum())
```

##  Encode the Target Label

The `species` column is a string category. We convert it to integers (0, 1, 2) so sklearn models can work with it.

```python
le = LabelEncoder()
iris['species'] = le.fit_transform(iris['species'])

# Show the mapping so it's clear what each number means
print("Label mapping:", dict(zip(le.classes_, le.transform(le.classes_))))
iris.head()
```

##  Normalise Features

Min-Max scaling brings all features into [0, 1]. This prevents features with large ranges (like petal length) from dominating distance-based models.

```python
X = iris.drop('species', axis=1)
y = iris['species']

scaler = MinMaxScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

print("Feature ranges after scaling:")
print(X_scaled.describe().loc[['min', 'max']])
```

##  Train / Test Split

We reserve 20 % of the data for evaluation. Setting `random_state` makes the split reproducible.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

print(f"Training samples : {len(X_train)}")
print(f"Test samples     : {len(X_test)}")
```
